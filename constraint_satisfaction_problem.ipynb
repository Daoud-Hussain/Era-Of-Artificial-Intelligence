```python
def solve_queens():
    board = [['_' for _ in range(4)] for _ in range(4)]
    queens = []

    def is_valid(row, col):
        for q in queens:
            r, c = q
            if row == r or col == c or abs(row - r) == abs(col - c):
                return False
        return True

    def print_board(board):
        for row in board:
            print(' '.join(row))

    for i in range(4):
        while True:
            pos = input(f"Enter the starting position for queen {i + 1} (e.g., a1, b2, etc.): ") 
            col = ord(pos[0].lower()) - ord('a')
            row = int(pos[1]) - 1
            if 0 <= row < 4 and 0 <= col < 4:
                if board[row][col] == '_':
                    queens.append((row, col))
                    board[row][col] = 'Q'
                    break
                else:
                    print("Invalid position. Try again.")
            else:
                print("Invalid position. Try again.")

    print("\nInitial Chessboard Configuration:")
    print_board(board)

    def place_queens():
        for row in range(4):
            for col in range(4):
                if is_valid(row, col):
                    queens.append((row, col))
                    board[row][col] = 'Q'
                    if len(queens) == 4:
                        return True
                    if place_queens():
                        return True
                    queens.pop()
                    board[row][col] = '_'
        return False

    queens.clear()
    board = [['_' for _ in range(4)] for _ in range(4)]
    if place_queens():
        print("\nFinal Chessboard Configuration (Queens Placed Correctly):")
        print_board(board)
    else:
        print("\nNo solution found.")

# Calling the method to solve and print the solution
solve_queens()
```

```python
def create_australia_graph():
    australia_graph = {
        'Western Australia': ['Northern Territory', 'South Australia'],
        'Northern Territory': ['Western Australia', 'South Australia', 'Queensland'],
        'South Australia': ['Western Australia', 'Northern Territory', 'Queensland', 'New South Wales', 'Victoria'],
        'Queensland': ['Northern Territory', 'South Australia', 'New South Wales'],
        'New South Wales': ['Queensland', 'South Australia', 'Victoria'],
        'Victoria': ['South Australia', 'New South Wales'],
        'Tasmania': []
    }

    return australia_graph

def color_australia():
    # Get the map of Australia
    australia_graph = create_australia_graph()

    # Define the available colors
    colors = ['Red', 'Green', 'Blue']

    # Initialize a dictionary to store the region-color assignments
    coloring = {}

    # Define the recursive backtracking function
    def backtrack(region):
        # Check if all regions have been assigned a color
        if len(coloring) == len(australia_graph):
            return True

        # Try assigning colors to the current region
        for color in colors:
            valid_assignment = True

            # Check if the color is valid for neighboring regions
            for neighbor in australia_graph[region]:
                if neighbor in coloring and coloring[neighbor] == color:
                    valid_assignment = False
                    break

            if valid_assignment:
                coloring[region] = color

                # Move to the next region
                next_region = next((r for r in australia_graph if r not in coloring), None)
                if next_region is None or backtrack(next_region):
                    return True

                # Backtrack if no valid assignment is found
                del coloring[region]

        return False

    # Start the backtracking algorithm
    if backtrack('Western Australia'):
        # Print the coloring of the regions
        for region, color in coloring.items():
            print(f'{region}: {color}')
    else:
        print('No solution found.')

# Solve and print the coloring of Australia
color_australia()
```
